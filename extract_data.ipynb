```python
import json

with open('./data/code_summarization_EN.json', 'r') as file:
    data = json.load(file)

data  # Now this is a list of dicts, one per line in your .jsonl file
```

```python
# take data and transform into pandas dataframe
import pandas as pd
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# iterate through the df and save only the input as a separate txt file for now
# get the first 15 rows
df = df.head(15)
import os
os.makedirs('./data/txt_files', exist_ok=True)
for i, row in df.iterrows():
    with open(f'./data/txt_files/row_{i}.txt', 'w') as file:
        file.write(f"Code:\n{row['input']}\n")
```
