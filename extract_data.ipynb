```python
from datasets import load_dataset
import pandas as pd
```

```python
dataset = load_dataset("iidai/codenet")
```

```python
df = dataset["train"].to_pandas()
df.head()
```

```python
# Extract code from df and save to individual files, files are named by language and id
# File extension is depending on the language
import os
os.makedirs("code_files", exist_ok=True)

for index, row in df.iterrows():
    lang = row["language"].lower()
    code = row["code"]
    id = row["id"]
    if lang == "python":
        ext = "py"
    elif lang == "java":
        ext = "java"
    elif lang == "cpp":
        ext = "cpp"
    elif lang == "c":
        ext = "c"
    elif lang == "javascript":
        ext = "js"
    elif lang == "ruby":
        ext = "rb"
    elif lang == "go":
        ext = "go"
    elif lang == "php":
        ext = "php"
    elif lang == "swift":
        ext = "swift"
    elif lang == "scala":
        ext = "scala"
    elif lang == "kotlin":
        ext = "kt"
    else:
        ext = "txt"  # default to txt if language is unknown
    filename = f"code_files/{lang}_{id}.{ext}"
    with open(filename, "w") as f:
        f.write(code)
```
